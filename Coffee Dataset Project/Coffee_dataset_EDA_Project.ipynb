```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.shape
```

#### 3) Depict different Market types and their count on a bar chart

```python
d3 = df['Market'].value_counts()
d3
```

```python
print(d3.index)
print(d3.values)
print(d3.index)
```

```python
plt.bar(d3.index,d3.values)
plt.show()
```

#### 4) Depict different Product types and their count on a line chart

```python
d4 = df['Product'].value_counts()
d4
```

```python
plt.plot(d4.index,d4.values,marker='d',color='maroon')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
d4b = df['Product Type'].value_counts()
d4b
```

```python
plt.plot(d4b.index,d4b.values,marker='d',color='maroon')
# plt.xticks(rotation=90)
plt.grid()
plt.show()
```

#### 5) Depict different State and their count on a pie chart

```python
d5 = df['State'].value_counts()
d5
```

```python
plt.figure(figsize=(8,8))
plt.pie(x =d5.values, labels=d5.index,autopct='%.2f%%')
plt.show()
```

#### 6) Depict Market wise sum of Profit and depict the result on bar chart.

```python
d6 = df.groupby(['Market'])['Profit'].sum()
d6
```

```python
plt.bar(d6.index,d6.values)
plt.show()
```

#### 7) Depict Product wise mean of Sales and depict the result on bar chart and line chart superimposed

```python
df.columns
```

```python
d7 = df.groupby(['Product'])['Sales'].mean()
d7
```

```python
plt.bar(d7.index,d7.values,color='lightblue',label='Mean Sales -Bar Chart')
plt.plot(d7.index,d7.values,color='maroon',label='Mean Sales - Line Chart',marker='*')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

#### 8) Depict Profit vs Sales on a Scatter Chart where Product is 'Regular Espresso'

#### 9) Depict Budget Profit vs Budget Sales on a Scatter Chart where State is California and Product is 'Earl Grey'

#### 10)  Depict Market wise mean of Budget Sales and mean of Sales in a stacked bar chart

```python
d10a = df.groupby(['Market'])['Budget Sales'].mean()
d10b = df.groupby(['Market'])['Sales'].mean()
d10a
```

```python
d10b
```

```python
plt.bar(d10a.index,d10a.values,color='grey',label='Budget Sales Mean')
plt.bar(d10b.index,d10b.values,color='lightblue',label='Sales Mean',alpha=0.5)
plt.legend(loc=10)
plt.show()
```

```python
d10 = df.groupby(['Market'])[['Budget Sales','Sales']].mean()
d10
```

```python
plt.bar(d10.index,d10['Budget Sales'],color='grey',label='Budget Sales Mean')
plt.bar(d10.index,d10['Sales'],color='lightblue',label='Sales Mean',alpha=0.5)
plt.legend(loc=10)
plt.show()
```

#### 26) Depict Market and Product Type based mean of Sales on a bar chart and on a stacked bar chart separately

```python
d26 = df.groupby(['Market','Product Type'])['Sales'].mean()
d26
```

```python
d26.plot(kind='bar')
plt.show()
```

```python
d26.unstack()
```

```python
d26.unstack().plot(kind='bar',stacked=True)
plt.grid()
plt.show()
```
